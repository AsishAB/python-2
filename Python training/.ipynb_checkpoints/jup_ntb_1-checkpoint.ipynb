```python
#Player with the maximum sixes
df.sort_values(by = ["SIXERS"], ascending = False)[:1]["PLAYER NAME"]
```

```python
#Player with the maximum sixes --> Another way
df.sort_values(by = ["SIXERS"], ascending = False)["PLAYER NAME"].iloc[0,0]
```

```python
#Most expensive player
df.sort_values(by = ["SOLD PRICE"], ascending = False)[:1]["PLAYER NAME"]
```

```python
#Player , sold price and team (top 5)
df[["PLAYER NAME", "SOLD PRICE", "TEAM"]].sort_values(by=["SOLD PRICE"], ascending=False).iloc[:5]
```

```python
# Australian player with highest ODI Wickets
df[df["COUNTRY"] == "AUS"].sort_values(by=["ODI-WKTS"], ascending=False)[["PLAYER NAME"]].iloc[0,0]
```

```python
# West Indian player with highest ODI Run
df[df["COUNTRY"] == "WI"].sort_values(by=["ODI-RUNS-S"], ascending=False)[["PLAYER NAME"]].iloc[0,0]
```

```python
# ALl countries
country_list = df["COUNTRY"]
country_list
```

```python
#Print total number of unique countries
countries = []
for i in country_list:
    if i not in countries:
        countries.append(i)

countries    
print(len(countries))
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
sn.countplot(x = "COUNTRY", data = df)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sn.countplot(x= "COUNTRY", data = df, ax = ax)
plt.title("COUNT VS COUNTRY", fontsize = 24)
plt.savefig('player-count-vs-country.jpg')
```

```python
fig, ax = plt.subplots(figsize = (12, 30))
df['COUNTRY'].value_counts().plot.pie()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12, 30))
df['COUNTRY'].value_counts().plot.pie(autopct = '%1.1f%')
plt.show()
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header=None)
df
```

```python
df.columns = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Flower']
df 
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sn.countplot(x= "Flower", data = df, ax = ax)
plt.title("COUNT VS Flowers", fontsize = 24)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
df['Flower'].value_counts().plot.pie()
plt.show()
```
